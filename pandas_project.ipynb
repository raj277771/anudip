```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/vgsales.csv')

# Group by 'Year' and sum 'Global_Sales'
yearly_sales = df.groupby('Year')['Global_Sales'].sum()

# Plotting the line plot
plt.figure(figsize=(10, 6))
plt.plot(yearly_sales.index, yearly_sales.values, marker='o')
plt.title('Global Sales Over the Years')
plt.xlabel('Year')
plt.ylabel('Global Sales (in millions)')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/vgsales.csv')

# Group by 'Publisher' and sum 'Global_Sales'
publisher_sales = df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False).head(10)  # Top 10 publishers

# Plotting the bar plot
plt.figure(figsize=(12, 6))
publisher_sales.plot(kind='bar')
plt.title('Top 10 Publishers by Total Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (in millions)')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/vgsales.csv')

# Plotting the box plot
plt.figure(figsize=(10, 6))
df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].boxplot()
plt.title('Box Plot of Sales in Different Regions')
plt.xlabel('Region')
plt.ylabel('Sales (in millions)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/vgsales.csv')

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['NA_Sales'], df['EU_Sales'])
plt.title('Scatter Plot of NA Sales vs. EU Sales')
plt.xlabel('NA Sales (in millions)')
plt.ylabel('EU Sales (in millions)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/vgsales.csv')

# Group by 'Genre' and sum 'Global_Sales'
genre_sales = df.groupby('Genre')['Global_Sales'].sum()

# Plotting the pie chart
plt.figure(figsize=(10, 6))
genre_sales.plot(kind='pie', autopct='%1.1f%%')
plt.title('Proportion of Global Sales by Genre')
plt.ylabel('')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/vgsales.csv')

# Selecting the relevant columns for the heatmap
sales_data = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]

# Calculating the correlation matrix
corr_matrix = sales_data.corr()

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Sales Regions and Global Sales')
plt.show()
```
